```python
# Importing dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from config import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "Output/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
# len(cities)
cities_df = pd.DataFrame(cities)
cities_df = cities_df.rename(columns={0: 'City'})
cities_df.head()
```

```python
# Perform a weather check on each city using a series of successive API calls
url = "http://api.openweathermap.org/data/2.5/weather?"

# To import temperature in Fahrenheit
units = "imperial"

city_names = []
cloudines = []
countries = []
dates = []
humidities = []
lat = []
lng = []
max_temp = []
wind_speed = []

# index counter
index = 0

for city in cities:
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units
    time.sleep(1)
    response = requests.get(query_url)
    weather_response = response.json()
        
    # Include a print log of each city as it's being processed (with the city number and city name).
    
    print("Processing record # " + str(index) + ": " + city)
    index += 1

    try:
        city = weather_response['name']
        cloudiness = weather_response['clouds']['all']
        country = weather_response['sys']['country']
        date = weather_response['dt']
        humidity = weather_response['main']['humidity']
        lats = weather_response['coord']['lat']
        lngs = weather_response['coord']['lon']
        mx_temp = weather_response['main']['temp_max']
        wind_sp = weather_response['wind']['speed']

        city_names.append(city)
        cloudines.append(cloudiness)
        countries.append(country)
        dates.append(date)
        humidities.append(humidity)
        lat.append(lats)
        lng.append(lngs)
        max_temp.append(mx_temp)
        wind_speed.append(wind_sp)
        print("Retrieving Data.")

    except:
        print("City data not found. Skipping city")
        continue

summary = {'City': city_names,
          'Cloudiness (%)': cloudines,
          'Country': countries,
          'Date': dates,
          'Humidity': humidities,
          'Latitude': lat,
          'Longitude': lng,
          'Max Temp': max_temp,
          'Wind Speed': wind_speed}
city_temp_df = pd.DataFrame(summary)
city_temp_df.head()
```

```python
# Exporting output to csv file
city_temp_df = city_temp_df.set_index(city_temp_df['City'])
city_temp_df.to_csv(output_data_file,index=False)
```

```python
# Importing csv data into another dataframe to avoid running the above lengthy code multiple times
city_temp_df2 = pd.read_csv(output_data_file)
city_temp_df2.head()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames

north_hem_df = city_temp_df2.loc[city_temp_df2['Latitude'] >= 0]
# north_hem_df.head()
south_hem_df = city_temp_df2.loc[city_temp_df2['Latitude'] < 0]
# south_hem_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hem_df['Latitude']
y_values = north_hem_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
scatterplot = plt.scatter(x=x_values,
                          y=y_values,
                          marker="o",
                          facecolors="cornflowerblue",
                          edgecolors="cornflowerblue",
                          alpha=0.75)
plt.title(f"Northern Hemisphere - Temperature (F) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5,5),fontsize=14,color="red")
print(f"The r-squared is: {round(rvalue,4)}")
plt.grid(alpha=0.15)
plt.tight_layout()
plt.savefig('Output/NorthHem_lat_temp_plot.png')
plt.show()
```

```python
## In the northern hemisphere, temperature seems to decrease as one moves up the latitude- a strong negative correlation
## This is exhibited in the plot above with all the latitudes, as well as by the correlation coefficient (r-squared): -0.89
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = south_hem_df['Latitude']
y_values = south_hem_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
scatterplot = plt.scatter(x=x_values,
                          y=y_values,
                          marker="o",
                          facecolors="cornflowerblue",
                          edgecolors="cornflowerblue",
                          alpha=0.75)
plt.title(f"Southern Hemisphere - Temperature (F) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(-55,88),fontsize=14,color="red")
print(f"The r-squared is: {round(rvalue,4)}")
plt.grid(alpha=0.15)
plt.savefig('Output/SouthHem_lat_temp_plot.png')
plt.show()
```

```python
## In the southern hemisphere, temperature seems to increase as one moves up the latitude, towards the Equator
## There appears to be a moderate positive correlation as exhibited by the correlation coefficient (r-squared): 0.48
## This is exhibited in the plot above with all the latitudes as well
## General Observation: Temperature increases nearer to the Equator, and decreases as we move away from it on either sides
```
